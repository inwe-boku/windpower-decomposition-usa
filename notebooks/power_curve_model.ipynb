# Power curve model

Some plots from the power curve model by Ryberg et al.

```python
from init import *
```

```python
matplotlib.rc('figure', figsize=(15, 10))
```

```python
power_curves = load_power_curve_model()
```

```python
def plot_power_curves(max_wind_speed=None):
    for specific_power in power_curves.specific_power[::3]:
        (100 * power_curves).sel(specific_power=specific_power, wind_speeds=slice(None, max_wind_speed)).plot(
            label=f"Specific power = {float(specific_power):.2f}W/mÂ²"
        )

    plt.legend();
    plt.ylabel('Capacity factor (%)')
    plt.xlabel('Wind speed (m/s)')
    plt.title('');
```

```python
plot_power_curves()
```

Note that an additional data point 0% capacity factor at 0m/s wind speed has to be added:

```python
plot_power_curves(max_wind_speed=5)
```
